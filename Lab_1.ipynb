Написать генератор случайных матриц(многомерных), который принимает
опциональные параметры <b>m</b>, <b>n</b>, <b>min_limit</b>, <b>max_limit</b>, где <b>m</b> и <b>n</b> указывают размер

```python
import random
import time
import numpy as np
```

```python
# Вводим параметры
user_m = int(input())
user_n = int(input())
user_min_limit = int(input())
user_max_limit = int(input())
# Генерируем матрицу в зависимости от параметров
def Matrix(user_m, user_n, user_min_limit, user_max_limit):
    # Создаем одномерный массив, который будет содержать другой массив - строки матрицы
    matrix2d = []
    # В каждую ячейку добавляем по массиву
    for i in range(user_m):
        matrix1d = []
        # Заполняем строки матрицы случайнвми числами
        for j in range(user_n):
            matrix1d.append(random.randint(user_min_limit, user_max_limit))
        matrix2d.append(matrix1d)
    return matrix2d
print(np.array(Matrix(user_m, user_n, user_min_limit, user_max_limit)))
```

```python
# Сортировка обменом. || Сортировка пузырьком.
start_time = time.time()
# Сгенерируем матрицу
unsortArr = Matrix(10,10,0,1000)
print("неотсортированная матрица: ")
print(np.array(unsortArr))
# Напишем функцию сортировки для одномерного массива
def booble_sort(array):
    # Создадим переменную для длины массива, которую будем изменять
    length = len(array)
    while(length != 0):
        # Запустим цикл, который за свою итерацию поставит максимальный элемент
        # на свое последнее место
        for i in range(1, length):
            if array[i] < array[i-1]:
                temp = array[i-1]
                array[i-1] = array[i]
                array[i] = temp
        length -= 1
    return array   
# Напишем функцию сортировки матрицы
def matrix_sort(array2d):
    for i in range(0, len(array2d)):
        booble_sort(array2d[i])
    return array2d
# Выведем отсортированную матрицу    
print("отсортированная матрица: ")
print(np.array(matrix_sort(unsortArr)))
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка выбором.
start_time = time.time()
# Сгенерируем матрицу для которой проведем сортировку
unsortArr = Matrix(10,10,0,1000)
print("неотсортированная матрица: ")
print(np.array(unsortArr))
# Напишем функцию сортировки для одномерного массива
def selection_sort(array):
    # Будем проходить по элементам массива, запоминая минимальным индекс первого элемента
    # в неотсортированной части массива. В результате полного перебора последний элемент
    # будет на своем месте поэтому его не включаем
    for i in range(0, len(array)-1):
        min_index = i
        # внутрениим циклом проходя по неотсортированной части массива, определим индекс
        # минимального элемента
        for j in range(i+1, len(array)):
            if array[j] < array[min_index]:
                min_index = j
        # Если первый неотсортированный элемент так и остался минимальным, то он стоит
        # на своем месте и нужно лишь сдвинуть цикл. Если нашелся другой минимальный элемент
        # то поменяем элементы местами
        if (min_index != i):
            temp = array[min_index]
            array[min_index] = array[i]
            array[i] = temp
    return array
# Напишем функцию сортировки матрицы
def matrix_sort(array2d):
    for i in range(0, len(array2d)):
        selection_sort(array2d[i])
    return array2d
# Выведем отсортированную матрицу    
print("отсортированная матрица: ")
print(np.array(matrix_sort(unsortArr)))
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка вставкой.
start_time = time.time()
# Сгенерируем матрицу
unsortArr = Matrix(10,10,0,1000)
print("неотсортированная матрица: ")
print(np.array(unsortArr))
# Напишем функцию сортировки для одномерного массива
def insertion_sort(array):
    # Предположим, что первый элемент массива отсортирован, пройдемся в цикле по
    # остальной неотсортированной части массива
    for i in range(1, len(array)):
        # создадим переменную, показывающую на каком элементе кончается
        # отсортированная часть массива
        sorted_arr = i-1
        # Если максимальный элемент отсортированной части больше следуещего
        # и в отсортированной части элементы не отсортированы вставляем элемент
        # в отсортированную часть на свое место
        while array[sorted_arr] > array[sorted_arr+1] and sorted_arr > -1:
            # меняем последний отсортированный местами со следующим неотсортированным
            temp = array[sorted_arr]
            array[sorted_arr] = array[sorted_arr+1]
            array[sorted_arr+1] = temp
            # внутри отсортированной части проверяем, стоит ли на своем месте введенный элемент
            sorted_arr -= 1
    return array    
# Напишем функцию сортировки матрицы
def matrix_sort(array2d):
    for i in range(0, len(array2d)):
        insertion_sort(array2d[i])
    return array2d
# Выведем отсортированную матрицу    
print("отсортированная матрица: ")
print(np.array(matrix_sort(unsortArr)))
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка Шелла.
start_time = time.time()
# Сгенерируем матрицу
unsortArr = Matrix(10,10,0,1000)
print("неотсортированная матрица: ")
print(np.array(unsortArr))
# Напишем функцию сортировки для одномерного массива
def shell_sort(array):
    # Выберем первоначальное расстояние между элементами для сортировки
    step = len(array)//2
    # Запустим цикл для сортировки с изменением шага
    while step > 0:
        # Внутри шагового цикла запустим стандартную сортировку вставками
        for i in range(step, len(array)):
            sorted_arr = i
            while array[sorted_arr - step] > array[sorted_arr] and sorted_arr >= step:
                temp = array[sorted_arr - step]
                array[sorted_arr - step] = array[sorted_arr]
                array[sorted_arr] = temp
                sorted_arr -= step
        # уменьшим шаг
        step //= 2
    return array    
# Напишем функцию сортировки матрицы
def matrix_sort(array2d):
    for i in range(0, len(array2d)):
        shell_sort(array2d[i])
    return array2d
# Выведем отсортированную матрицу    
print("отсортированная матрица: ")
print(np.array(matrix_sort(unsortArr)))
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Быстрая сортировка.
start_time = time.time()
# Сгенерируем матрицу
unsortArr = Matrix(10,10,0,1000)
print("неотсортированная матрица: ")
print(np.array(unsortArr))
# Сортировка Хоара состоит из нескольких функций, напишем их
# Функция partition является ядром сортировки, она принимает массив и два указателя
# и постепенно сдвигая указатели создает два отсортированных относительно друг друга
# подмассива относительно опорного, все элементы слева от опорного меньше любого правого
def partition(array, left, right):
    # Создадим переменную разделитель, отвечающую за опорный элемент, выберем последний элемент
    pivot = array[right]
    # Для удобства переобозначим элементы и раз последий элемент опорный, сдвинем правый указатель
    i = left
    j = right - 1
    # пока указатели не встретятся запустим цикл
    while i < j:
        # будем сдвигать указатели к опорному, если они на своих местах и их не нужно переставлять
        while array[i] < pivot and i < right:
            i +=1
        while (array[j] >= pivot and j > left):
            j -=1
        # дойдя до элементов которые не соответсвуют своим областям
        # где элемент слева больше опорного, а справа меньше. Если при этом
        # указатели не встретились, поменяем их местами и сдвинем указатели
        if i < j:
            array[i], array[j] = array[j], array[i]
            i +=1
            j -=1
    # После отсортировывания элементов относительно опорного,
    # поставим опорный на нужное место
    if array[i] > pivot:
        array[i], array[right] = array[right], array[i]
    # вернем индекс элемента относительно которого элементы слева меньше любого элемента справа
    return i
# Напишем функцию самой сортировки, которая будет представлена двумя рекрусиями
def quick_sort(array, start, end):
    if start < end:
        right_start = partition(array, start, end)
        quick_sort(array, start, right_start - 1)
        quick_sort(array, right_start + 1 , end)
# Напишем функцию сортировки матрицы
def matrix_sort(array2d):
    for i in range(0, len(array2d)):
        quick_sort(array2d[i], 0, len(array2d[i])-1)
    return array2d
# Выведем отсортированную матрицу    
print("отсортированная матрица: ")
print(np.array(matrix_sort(unsortArr)))
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
import heapq
# Турнирная сортировка.
start_time = time.time()
# Для турнирной сортировки необходимо использовать маленькую трех-уровневую кучу из 7 элементов
# эта куча будет принимать в себя массив и распологать данные из него в минимально древовидном порядке
# где на самом нижнем уровне самый минимальный элемент и каждый родитель меньше потомка

# Вспомогательная функция для объединения отсортированных массивов
def merge_arrays(arrayOfArrays):
    # Если массив один, преобразуем его из двумерного в одномерный
    if len(arrayOfArrays) == 1:
        return arrayOfArrays[0]
    # Если массивов несколько, создадим результирующий и отсортируем массивы по убыванию
    result = []
    arrayOfArrays.sort(reverse=True)
    # Будем в цикле перебирать два одномерных массива, формируя один отсортированный, который
    # затем пойдет на следующую итерацию цикла для объединения со следующим и т.д.
    for i in range(0, len(arrayOfArrays)-1):
        while arrayOfArrays[i] and arrayOfArrays[i+1]:
            # Пока массивы не пусты, будем вынимать первый элемент в соответствии с условием
            if arrayOfArrays[i][0] < arrayOfArrays[i+1][0]:
                result.append(arrayOfArrays[i].pop(0))
            else:
                result.append(arrayOfArrays[i+1].pop(0))
        # Если по итогу опустел i-ый массив, то по выходу из цикла остался не записанный в результат
        # i+1 массив, добавим его к результату, иначе добавим i-ый
        if not arrayOfArrays[i]:
            result += arrayOfArrays[i+1]
        else:
            result += arrayOfArrays[i]
        arrayOfArrays[i+1] = result
    return result

def tournament_sort(arr):
    # Создадим массив, который будет содержать отсортированные массивы победителей
    all_winners = []
    flag = True
    while flag:
        # Создадим массив данных, состоящий из 7 элементов, куда занесем первые 7 элементов массива
        data = []
        if len(arr) >= 7:
            for i in range(0, 7):
                data.append(arr[i])
            i = 7
        # Если исходный массив содержит меньше чем 7 элементов, то просто выгрузим что есть в данные
        else:
            data = arr
            i = len(data)+1
        # Прервратим данные в кучу, которая по умолчанию выведет минимальный элемент на нижний уровень
        heapq.heapify(data)
        # Создадим массив проигравших и массив победителей
        losers = []
        winners = []
        # Занесем первый минимальный элемент из кучи в массив победителей, при этам после изъятия
        # элемента из кучи, куча переформируется сохранив древовидную структуру
        winners.append(heapq.heappop(data))
        # Будем пополнять кучу пока не закончится сортируемый массив
        while i < len(arr):
            # Сравним последний элемент массива победителей с элементом который собираемся внести в кучу
            # Если он больше, то добавим его в кучу, если равен, то внесем в победители, иначе в проигравшие
            if arr[i] > winners[-1]:
                heapq.heappush(data, arr[i])
                # Если мы внесли элемент в кучу, то сразу достанем из нее нижний элемент и внесем в победители
                winners.append(heapq.heappop(data))
            elif arr[i] == winners[-1]:
                winners.append(arr[i])
            else:
                losers.append(arr[i])
            i +=1
        # По итогам итерации, мы перенесли все элементы в массивы победителей и проигравших
        # Занесем оставшиеся в куче элементы в победителей
        while data:
            winners.append(heapq.heappop(data))
        
        flag = False
        # Добавим партию победителей в массив всех победителей
        all_winners.append(winners)
        # Если у нас есть проигравшие, то для них запустим еще одну итерацию сортировки
        if losers:
            arr = losers
            flag = True
    # Получив отосртированные массивы победителей, объединим их сохранив сортировку
    return merge_arrays(all_winners)
       
        
# Функция сортировки матрицы
def matrix_sort(array2d):
    array2d2 = []
    for i in range(0, len(array2d)):
        array2d2.append(tournament_sort(array2d[i]))
    return array2d2
# Сгенерируем матрицу
unsortArr = Matrix(10,10,0,1000)
print("неотсортированная матрица: ")
print(np.array(unsortArr))
# Выведем отсортированную матрицу    
print("отсортированная матрица: ")
print(np.array(matrix_sort(unsortArr)))
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```
