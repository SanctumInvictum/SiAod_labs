```python
# Линейный поиск
def LinearSearch(arr, element):
    for i in range (len(arr)):
        if arr[i] == element:
            return i
    return None

arr = [random.randrange(1000) for i in range(100)]
print('Сгенерируем рандомный массив для поиска:\n',arr)
target = int(input())
start_time = time.time()
result = LinearSearch(arr,target)
if result is not None:
    print('\nPosition: ',result)
else:
    print('Не найдено')
print("-----Линейный поиск затратил {0} ms-----".format((time.time() - start_time)*1000))
```

```python
# Бинарный поиск
# Итеративная реализация
def BinarySearch(arr, element):
    # обозначим элементы между которыми находится искомый
    first = 0
    last = len(arr)-1
    # Сдвигаем границы для поиска 
    while first <= last:
        # Делим пополам
        mid = int((first + last)/2)
        # Условие выхода
        if arr[mid] == element:
            return mid
        # Если меньше среднего, сдвигаем правую границу
        elif element < arr[mid] :
            last = mid-1
        # Иначе левую
        else:
            first = mid+1
    return None

arr = [random.randrange(1000) for i in range(100)]
arr.sort()
print('Сгенерируем и отсортируем рандомный массив:\n',arr)

target = int(input())
start_time = time.time()
result = BinarySearch(arr,target)
if result is not None:
    print('Искомый индекс: ',result)
else:
    print('Не найдено')
print("-----Бинарный поиск затратил {0} ms-----".format((time.time() - start_time)*1000))
# Временная сложность O(log n)
```

```python
# Бинарное дерево поиска
class Node:
    # Метод инициализации узла
    def __init__(self, input):
        self.left = None
        self.right = None
        self.value = input
# Древо поиска
class BinarySearchTree:
    # Инициализация корня
    def __init__(self):
        self.root = None
    
    # Метод добавления элемента в древо
    def insert(self, input):
        # Добавляем корень
        if self.root is None:
            self.root = Node(input)
        else:
        # Добавляем узел
            self.internal_insert(input, self.root)
    
    # Метод добавления узла внутри древа
    def internal_insert(self, input, node):
        # Если меньше корня, то добавляем в левую ветвь
        if input < node.value:
            if node.left is None:
                node.left = Node(input)
            else:
                self.internal_insert(input, node.left)
        # Если больше, то в правую ветвь
        else:
            if node.right is None:
                node.right = Node(input)
            else:
                self.internal_insert(input, node.right)
    
    # Внешний поиск
    def search(self, input):
        return self.internal_search(input, self.root)
    
    # Поиск рекурсией внутри дерева
    def internal_search(self, input, node):
        # Если нашли элемент или если искомого элемента нет, возвращаем индекс или None
        if node is None or node.value == input:
            return node
        # в зависимости от значения запускаем поиск в соответствующей ветви
        elif input < node.value:
            return self.internal_search(input, node.left)
        else:
            return self.internal_search(input, node.right)
        
arr = [random.randrange(1000) for i in range(100)]
print('Сгенерируем рандомный массив:\n',arr)
# Создадим дерево и заполним его элементами из массива
tree = BinarySearchTree()
for num in arr:
    tree.insert(num)

target = int(input())
start_time = time.time()
result = tree.search(target)
if result is not None:
    print('Искомое число присутствует в списке')
else:
    print('Не найдено')
print("-----Древесный поиск затратил {0} ms-----".format((time.time() - start_time)*1000))
```

```python
# Фибоначчиев поиск
def FibonacciSearch(arr, element):
    # Задаем два фибоначиевых числа и третий в виде суммы предыдущих
    fib1, fib2 = 0, 1
    fib3 = fib1 + fib2
    # увеличиваем тройку пока наше старшее фибоначиево число не станет больше числа элементов массива
    while (fib3 < len(arr)):
        fib1 = fib2
        fib2 = fib3
        fib3 = fib1 + fib2
    # создаем нулевой индекс
    index = -1;
    # В цикле пока третье число больше базового третьего числа фибоначчи
    while (fib3 > 1):
        # Разделяем наш список на две части по первому числу из тройки
        i = min(index + fib1, (len(arr)-1))
        # Если искомое число находится после первого фибоначиевого числа из тройки
        # то уменьшаем тройку на 1 фибоначиево число, исключая левую часть
        if (arr[i] < element):
            fib3 = fib2
            fib2 = fib1
            fib1 = fib3 - fib2
            index = i
        # Если искомое число находится до первого фибоначиевого числа, уменьшаем на 2 числа,
        # исключая правую часть
        elif (arr[i] > element):
            fib3 = fib1
            fib2 = fib2 - fib1
            fib1 = fib3 - fib2
        else :
            # Возвращаем искомый индекс, если значение окажется равным 
            return i
    # Если мы не нашли число, значит оно либо второе, либо его нет
    if(fib2 == 1 and arr[index+1] == element):
        return index+1;
    return -1

arr = [random.randrange(1000) for i in range(100)]
arr.sort()
print('Сгенерируем и отсортируем рандомный массив:\n',arr)

target = int(input())
start_time = time.time()
result = FibonacciSearch(arr,target)
if result is not None:
    print('Искомый индекс: ',result)
else:
    print('Не найдено')
print("-----Фибоначиев поиск затратил {0} ms-----".format((time.time() - start_time)*1000))
```

```python
# Интерполяционный поиск
def InterpolationSearch(arr, element):
    # Выибраем границы внутри которых находится искомое число
    low = 0
    high = len(arr) - 1
    while low <= high and element >= arr[low] and element <= arr[high]:
        # Хитро формульным способом выбираем предполагаемое место искмого элемента
        pos = low + ((element - arr[low]) * (high - low) // (arr[high] - arr[low])) 
        # радуемся если угадали
        if arr[pos] == element:
            return pos
        # Грустим и если искомый элемент больше предположенного, чуть чуть увеличиваем нижнюю границу
        elif arr[pos] < element:
            low = pos + 1
        # Иначе также грустим и чутка уменьшаем верхнюю границу для уточнения поиска
        else:
            high = pos - 1
    # Совсем грустим если наше гадание не оправдалось
    return None

arr = [random.randrange(1000) for i in range(100)]
arr.sort()
print('Сгенерируем и отсортируем рандомный массив:\n',arr)

target = int(input())
start_time = time.time()
result = InterpolationSearch(arr,target)
if result is not None:
    print('Искомый индекс: ',result)
else:
    print('Не найдено')
print("-----Интерполяционный поиск затратил {0} ms-----".format((time.time() - start_time)*1000))
```

```python
# Псевдо-рандомное рехеширование
import random
class HashTable:
    def __init__(self, capacity):
        self.capacity = capacity
        self.table = [None] * capacity
        self.rand = random.randint(1, self.capacity - 1)
    
    def hash(self, key):
        return key % self.size
    
    def rehash(self, old_hash):
        return (old_hash + self.rand) % self.size
    
    def insert(self, key, value):
        hash_value = self.hash(key)
        if self.table[hash_value] is None:
            self.table[hash_value] = (key, value) 
        else:
            next_slot = self.rehash(hash_value)
            while self.table[next_slot] is not None and self.table[next_slot][0] != key:
                next_slot = self.rehash(next_slot)
            if self.table[next_slot] is None:
                self.table[next_slot] = (key, value)
            else:
                self.table[next_slot] = (key, value) # replace

    def get(self, key):
        hash_value = self.hash(key)
        if self.table[hash_value] is None:
            return None
        elif self.table[hash_value][0] == key:
            return self.table[hash_value][1]
        else:
            next_slot = self.rehash(hash_value)
            while self.table[next_slot] is not None and self.table[next_slot][0] != key:
                next_slot = self.rehash(next_slot)
            if self.table[next_slot] is None:
                return None
            elif self.table[next_slot][0] == key:
                return self.table[next_slot][1]
            else:
                return None
    

hash_table = HashTable(10)

hash_table.insert(54, "a")
hash_table.insert(26, "b")
hash_table.insert(93, "c")
hash_table.insert(17, "d")
hash_table.insert(77, "e")

print(hash_table.get(77))
print(hash_table.get(26))
```

```python
# Метод цепочек
# Класс узла нашего списка
class Node:
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.next = None
# Хэш-таблица
class HashTable:
    # создает таблицу с заданной начальной емкостью
    def __init__(self, capacity):
        self.capacity = capacity
        self.size = 0
        # для каждой ячейки создаем пустые списки
        self.buckets = [None] * self.capacity
    def insert(self, key, value):
        index = hash(key) % self.capacity
        node = self.buckets[index] # цепочка элементов по индексу
        while node: 
            # Если ключ есть, обновляем значением
            if node.key == key: 
                node.value = value
                return
            node = node.next
        # Иначе добавляем в начало цепочки
        new_node = Node(key, value)
        new_node.next = self.buckets[index]
        self.buckets[index] = new_node
        self.size += 1
        # Если элементов больше 80% от емкости, увеличиваем емкость 
        if self.size > 0.8 * self.capacity:
            self.resize()
    
    # значения по ключу
    def get(self, key):
        index = hash(key) % self.capacity
        node = self.buckets[index]
        while node:
            # Если нашли ключ найден, возвращаем значение
            if node.key == key:
                return node.value
            node = node.next
        return None
    
    # Метод удаления элемента
    def remove(self, key):
        index = hash(key) % self.capacity
        node = self.buckets[index]
        prev = None
        while node:
            # Если нашли нужный ключ, удаляем элемент
            if node.key == key: 
                if prev:
                    prev.next = node.next
                else:
                    self.buckets[index] = node.next
                self.size -= 1
                return
            prev = node
            node = node.next
    
    # Изменение емкости
    def resize(self):
        # увеличиваем вдвое
        new_capacity = self.capacity * 2
        new_buckets = [None] * new_capacity
        # рехешируем элементы в новую с измененной емкостью
        for i in range(self.capacity):
            node = self.buckets[i]
            while node:
                index = hash(node.key) % new_capacity
                if new_buckets[index]:
                    new_node = new_buckets[index]
                    while new_node.next:
                        new_node = new_node.next
                    new_node.next = Node(node.key, node.value)
                else:
                    new_buckets[index] = Node(node.key, node.value)
                node = node.next
        self.capacity = new_capacity
        self.buckets = new_buckets


table = HashTable(10)
table.insert('Apple iPhone 14', 500)
table.insert('Samsung Galaxy S22 Ultra', 300)
table.insert('Игровые приставки', 200)
table.insert('Philips Hue Smart Bulbs', 500)
#table.insert('key5', '5')
print(table.get('Apple iPhone 14'))
print(table.get('Samsung Galaxy S22 Ultra'))
print(table.get('Игровые приставки')) 
print(table.get('Philips Hue Smart Bulbs'))
#table.remove('key3')
#table.remove('key1')
#print(table.get('key1'))
#print(table.get('key2'))
#print(table.get('key3')) 
#print(table.get('key4')) 
#print(table.get('key5')) 
```
